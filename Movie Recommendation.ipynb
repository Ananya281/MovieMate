```python
def text_process(mess):
    nopunc = [char for char in mess if char not in string.punctuation]
    nopunc = ''.join(nopunc)
    return [word for word in nopunc.split() if word.lower() not in stopwords.words('english')]
```

```python
def recommend_based_actor(actor_name):
    # Normalize the actor_name by removing spaces and converting to lowercase
    actor_name_normalized = actor_name.replace(" ", "").lower()
    
    # Normalize the 'actors' column by removing spaces and converting to lowercase
    movienew['actors_normalized'] = movienew['actors'].str.replace(" ", "").str.lower()
    
    # Find the movies that contain the actor in the 'actors_normalized' column (partial match)
    matching_movies = movienew[movienew['actors_normalized'].str.contains(actor_name_normalized, na=False)]
    
    # If no movies found for the actor, return an empty list
    if matching_movies.empty:
        return []
    
    # Create a list to store the recommendations
    recommendations = []
    
    # Loop through the matching movies and find the top recommendations based on rating
    for index, row in matching_movies.iterrows():
        movie_info = [
            [row['title']],             # Movie title inside a list
            [row['imdb_rating']],       # Movie rating inside a list
            [row['poster_path']]        # Movie poster inside a list
        ]
        recommendations.append(movie_info)
    
    # Sort the recommendations by rating in descending order
    recommendations.sort(key=lambda x: x[1], reverse=True)
    
    # Limit the number of recommendations to the top 10
    return recommendations

def recommend_based_genre(genre_name):
    # Normalize the genre_name by removing spaces and converting to lowercase
    genre_name_normalized = genre_name.strip().lower()
    
    # Find movies that contain the exact genre name in the 'genres' column
    matching_movies = movienew[movienew['genres'].str.lower().str.contains(genre_name_normalized, case=False, na=False)]
    
    # If no movies are found for the genre, return an empty list
    if matching_movies.empty:
        return []
    
    # Create a list to store the recommendations
    recommendations = []
    
    # Loop through the matching movies
    for index, row in matching_movies.iterrows():
        movie_info = [
            [row['title']],             # Movie title inside a list
            [row['imdb_rating']],       # Movie rating inside a list
            [row['poster_path']]        # Movie poster inside a list
        ]
        recommendations.append(movie_info)
    
    # Sort the recommendations by rating in descending order
    recommendations.sort(key=lambda x: x[1], reverse=True)
    
    # Limit the number of recommendations to the top 10
    return recommendations

def recommend_based_movie(query):
    # Normalize the query
    query_normalized = query.strip().lower()
    
    # Normalize movie titles by removing extra text (e.g., '(film)')
    movie_titles_normalized = movienew['title'].str.lower().str.replace(r'\s*\(.*\)', '', regex=True)
    
    # Find the movies that match the query
    movie_matches = movienew[movie_titles_normalized.str.contains(query_normalized, na=False)]
    
    # If no movies are found, return an empty list
    if movie_matches.empty:
        return []
    
    # Create a list to store the recommendations
    recommendations = []
    
    # Loop through the matching movies
    for _, row in movie_matches.iterrows():
        try:
            # Find the index of the movie
            movie_index = movienew[movie_titles_normalized == row['title'].lower()].index[0]
            
            # Get similarity distances for that movie
            distances = similarity[movie_index]
            
            # Sort movies based on similarity and pick the top 5 (excluding the movie itself)
            movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
            
            for i in movies_list:
                movie_info = [
                    [movienew.iloc[i[0]]['title']],           # Movie title inside a list
                    [movienew.iloc[i[0]]['imdb_rating']],     # Movie rating inside a list
                    [movienew.iloc[i[0]]['poster_path']]      # Movie poster inside a list
                ]
                recommendations.append(movie_info)
        
        except IndexError:
            # Handle case where movie index is not found
            continue
    
    # Sort the recommendations by rating in descending order
    recommendations.sort(key=lambda x: x[1], reverse=True)
    
    return recommendations[:5]
```

```python
def combined_movie_recommendation(query):
    """
    Recommends movies based on the provided query (movie title, actor name, or genre).

    Args:
        query (str): The search query.

    Returns:
        list: A list of recommended movies, each containing title, rating, and poster path.
    """

    query_normalized = query.strip().lower()

    # Ensure 'actors_normalized' column exists (if not, create it)
    if 'actors_normalized' not in movienew.columns:
        movienew['actors_normalized'] = movienew['actors'].str.replace(" ", "").str.lower()

    # Check for movie title, actor, or genre match
    if query_normalized in movienew['title'].str.lower().values:
        return recommend_based_movie(query)[:5]
    elif query_normalized in movienew['actors_normalized'].values:
        return recommend_based_actor(query)[:5]
    elif query_normalized in movienew['genres'].str.lower().values:
        return recommend_based_genre(query)[:5]

    recommendations = []
    for func in [recommend_based_actor, recommend_based_genre, recommend_based_movie]:
        try:
            recs = func(query_normalized)
            recommendations.extend(recs)
        except:
            pass  # Handle potential errors in individual recommendation functions
    recommendations.sort(key=lambda x: x[1], reverse=True)
    return recommendations
```

```python
import pandas as pd

def recommend_movies(popular, selected_genres=None, selected_actors=None, top_n=None):
    """
    Recommends popular movies based on user preferences.

    Args:
        popular: The DataFrame containing movie data.
        selected_genres: A list of selected genres.
        selected_actors: A list of selected actors.
        top_n: The number of movies to recommend.

    Returns:
        A DataFrame of recommended movies.
    """

    if popular.empty:
        print("Error: The 'popular' DataFrame is empty.")
        return pd.DataFrame()

    # Check if 'imdb_rating' column exists
    if 'imdb_rating' not in popular.columns:
        print("Error: 'imdb_rating' column is missing in the DataFrame.")
        return pd.DataFrame()

    # Filtering based on selected genres (case-insensitive)
    if selected_genres:
        popular = popular[popular['genres'].apply(
            lambda x: any(genre.lower() in (item.lower() for item in x.split(',') if isinstance(x, str))
                          for genre in selected_genres)
        )]

    # Filtering based on selected actors (case-insensitive and allows partial matches)
    if selected_actors:
        popular = popular[popular['actors_normalized'].apply(
            lambda x: any(all(part.lower() in x.lower() for part in actor.split()) for actor in selected_actors)
        )]

    # Sorting by IMDb rating
    popular = popular.sort_values(by='imdb_rating', ascending=False)

    # Returning the top N recommendations
    return popular.head(top_n)

# Example Usage
reco = recommend_movies(popular, selected_genres=['Action'], selected_actors=['Kajol'], top_n=10)
print(reco[['title', 'imdb_rating']])
```

```python
from urllib.parse import unquote
```

```python
def get_movie_info(movienew_df, title):
    title = unquote(title)  # Decode URL-encoded title
    title_lower = title.lower()
    
    # Perform a case-insensitive search that allows partial matches
    movie_info = movienew_df[movienew_df['title'].str.lower().str.contains(title_lower, na=False)]
    
    if movie_info.empty:
        return None  # Return None if no matching title is found
    
    # Extract the first matching movie's details
    movie_details = {
        'title': movie_info['title'].values[0],
        'imdb_rating': movie_info['imdb_rating'].values[0],
        'poster_path': movie_info['poster_path'].values[0],
        'genres': movie_info['genres'].values[0],
        'summary': movie_info['summary'].values[0],
        'actors': movie_info['actors'].values[0],
    }
    return movie_details
```

```python
def get_movie_wiki(popular, title):
    # Filter the DataFrame for the specific movie title
    movie_info = popular[popular['title'] == title]
    
    if movie_info.empty:
        return None  # Return None if no matching title is found
    
    # Extract the story
    return movie_info['story'].values[0]
```
